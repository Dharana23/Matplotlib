```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('larger')
axes2.plot(y,x)
axes2.set_title('smaller')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('first')
axes[1].plot(y,x)
axes[1].set_title('second')

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

for current_ax in axes:
    current_ax.plot(x,y)

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (8,2),dpi = 100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (3,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('picture.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x square')
ax.plot(x,x**3, label='x cube')

ax.legend(loc=(0.1,0.1))
#ax.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',linewidth=3,alpha=0.5,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')   #linewidth or lw
#ax.plot(x,y,color='#FF8C00')
#alpha is for transparency
#linestyle or ls
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',linewidth=2,linestyle='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
